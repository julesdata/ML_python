## XLSX

```python
# Worldwide Mobile Data Pricing
# https://www.kaggle.com/nulldata/worldwide-mobile-data-pricing
# sheet_name=None : 모든 시트를 읽어옴. default는 '0' (첫번째 시트) 
df = pd.read_excel('global-mobile-data-price-comparison-2020.xlsx', None);
type(df)
```

```python
df.keys()  # sheet 명을 읽어옵니다. (df가 dict형태 이므로)
```

```python
# 하나의 sheet를 지정해서 DataFrame을 생성합니다.
df = df['2020 1GB Mobile data cost']
df.shape
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
# 필요 없는 컬럼 제거 방법 1 
target_columns=[column for column in df.columns if ':' not in column] #컬럼 이름 안에 ':'가 포함되지 않은 것
target_columns
```

```python
# 필요 없는 컬럼 제거 방법 2
target_column=df.columns.tolist()[:14]
target_column
```

```python
df=df[target_columns]
df.shape
```

```python
# 결측행 제거 
df = df[~df['Rank'].isnull()] #rank열에서 null이 아닌 행들만 df에 저장해줘
df.shape
```

```python
df.Name
```

```python
#'orea' 문자열을 포함한 name 컬럼값의 전체 행 조회 
df[df.Name.str.contains('orea')]
```

```python
#슬라이싱으로 조회 
df[df['Country code']=='KR']
```

```python
# 값이 빈 row를 제거합니다.
df = df[~df.Rank.isnull()]
df.shape
```

```python
df.tail()
```

```python
# 값이 빈 column을 제거합니다.
actual_columns = df.columns
actual_columns = [column for column in actual_columns if ':' not in column]
df = df[actual_columns]
df.keys()
```

```python
# 한국을 찾아봅니다.
df[df['Country code']=='KR']
```

```python
# 다른 국가도 찾아봅니다.
df[df.Name.str.contains('ndi')]
```

## JSON

중첩된 구조의 JSON 파일을 원하는 목적에 맞게 DataFrame으로 변환합니다.

```python
# https://www.kaggle.com/nyphil/perf-history
# with open('파일명','사용목적; w-write, r-read')
import json
with open('raw_nyc_phil.json', 'r') as f:   
    json_nyc = json.load(f)
type(json_nyc)
```

```python
json_nyc.keys()
```

```python
len(json_nyc['programs'])
```

```python
#json파일의 dataframe 파싱 매서드
#pd.jason_normalize (첫 key 값만 샘플로 보여준다) . r 에서는 json lite (전체 값을 펼침)
pd.json_normalize(json_nyc)
```

```python
json_nyc['programs'][0].keys()
```

```python
df=pd.json_normalize(json_nyc['programs'])
df
```

```python
# META를 지정하지 않고 record_path만 지정할 경우
df_without_meta = pd.json_normalize(json_nyc, ['programs'])
df_without_meta.shape
```

```python
df_without_meta.tail()
```

```python
#concerts 값을 보고싶을때
df_concerts=pd.json_normalize(json_nyc['programs'][0]['concerts'])
df_concerts
```

```python
#concerts는 리스트로 이루어져있고, 2개 이상일 수 도 있다. 
df_concerts=pd.json_normalize(json_nyc['programs'][1210]['concerts'])
df_concerts
```

```python
json_nyc['programs'][1210]['concerts'] 
```

```python
#concert 세부정보를, 상위정보와 함께 보고싶다면? 
json_nyc['programs'][1210]
```

```python
#  리스트가 여러개일 경우, season, orchestra등의 값은 중복되게 된다. list로 이루어져 있다. 중복 정보를 meta데이터라고 한다. 
# 특정 record_path내에서 meta를 지정. meta DataFrame을 따로 만들어 join 할 필요가 없음.
df_with_meta = pd.json_normalize(json_nyc['programs'], 
                                 ['concerts'], 
                                 meta=['season', 'orchestra', 'programID'])
df_with_meta.shape
```

```python
df_with_meta.tail()
```

```python
df_with_meta[df_with_meta.programID=='3281']
```
