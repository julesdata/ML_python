# Multiple Linear Regression


LinearRegression fits a linear model with coefficients w = (w1, …, wp) to minimize the residual sum of squares between the observed targets in the dataset, and the targets predicted by the linear approximation.

선형회귀분석. RSS(잔차 제곱 합) 값을 최소화 하는 방향으로 Y=wX 방정식의 w값을 구하는 모형.
잔차: y값과 y값으로 추정된 값의 차이

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html?highlight=linearregression#sklearn.linear_model.LinearRegression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Exercise
Graduate Admission 2

Context
- This dataset is created for prediction of Graduate Admissions from an Indian perspective.

Content
- The dataset contains several parameters which are considered important during the application for Masters Programs.
- The parameters included are :
    1. GRE Scores ( out of 340 )
    1. TOEFL Scores ( out of 120 )
    1. University Rating ( out of 5 )
    1. Statement of Purpose and Letter of Recommendation Strength ( out of 5 )
    1. Undergraduate GPA ( out of 10 )
    1. Research Experience ( either 0 or 1 )
    1. Chance of Admit ( ranging from 0 to 1 )

### Step1: Load Dataset

```python
# 'Admission_Predict.csv' 파일 읽기 pd.read_csv()
# df shape 확인

df = pd.read_csv('Admission_Predict.csv')
df.shape
```

```python
# df tail 행 보기

df.tail()
```

```python
# index역할을 하는 'Serial No.' 컬럼 제거하기

df = df.drop(['Serial No.'], axis = 1)
```

```python
# df descriptive statistics

df.describe()
```

### Step2: Visualize Dataset

```python
# sns.pairplot()

sns.pairplot(df)
```

```python
# corr matrix with heatmap
# sns.heatmap, df.corr() 활용

plt.figure(figsize = (8, 8))
sns.heatmap(df.corr(), annot = True, cmap = 'YlGnBu')
```

### Step3: Define Train set and Test Set

```python
# X에 -1번째 column까지 할당
# y에 -1번째 column을 할당

X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
# split train set and test set

# from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test 할당. test_size=0.2, random_state=1

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## with scikit-learn

### Step4: Train Model

```python
# define regressor type, and fit with train set

# from sklearn.linear_model import LinearRegression
# regressor 에 LinearRegression() 오브젝트 할당
# X_train, y_train 으로 regressor fit


from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# regression model's Estimated coefficients

# regressor 의 coefficient 확인

regressor.coef_
```

```python
# regression model's independent term in the linear model.
# Set to 0.0 if fit_intercept = False.

# regressor의 intercept 확인

regressor.intercept_
```

### Step5: Predict value

```python
# predict y values with trained regressor

y_pred = regressor.predict(X_test)
```

```python
# predict value visualization
plt.scatter(y_test, y_pred, color = 'r')
plt.ylabel('Model Predictions')
plt.xlabel('True (ground truth)')
```

### Step6: Evaluate Model

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from math import sqrt

k = X_test.shape[1]
n = len(X_test)

# RMSE(평균 제곱근 오차(Root Mean Square Error))
# https://en.wikipedia.org/wiki/Root-mean-square_deviation
RMSE = np.sqrt(mean_squared_error(y_test, y_pred))

MSE = mean_squared_error(y_test, y_pred)
MAE = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)
MAPE = np.mean( np.abs((y_test - y_pred) / y_test ) ) * 100

print('RMSE =',RMSE, 
      '\nMSE =',MSE, 
      '\nMAE =',MAE, 
      '\nR2 =', r2, 
      '\nAdjusted R2 =', adj_r2, 
      '\nMean Absolute Percentage Error =', MAPE, '%') 
```

```python
from scipy import stats
# reference. https://stackoverflow.com/questions/27928275/find-p-value-significance-in-scikit-learn-linearregression

params = np.append(regressor.intercept_, regressor.coef_)
predictions = regressor.predict(X_train)

newX = pd.DataFrame({"Constant":np.ones(len(X_train))}).join(pd.DataFrame(X_train.reset_index(drop=True)))
MSE = (sum((y_train-predictions)**2))/(len(newX)-len(newX.columns))

# Note if you don't want to use a DataFrame replace the two lines above with
# newX = np.append(np.ones((len(X_train),1)), X, axis=1)
# MSE = (sum((y_train-predictions)**2))/(len(newX)-len(newX[0]))

var_b = MSE*(np.linalg.inv(np.dot(newX.T,newX)).diagonal())
sd_b = np.sqrt(var_b)
ts_b = params/ sd_b

p_values =[2*(1-stats.t.cdf(np.abs(i),(len(newX)-len(newX.values[0])))) for i in ts_b]


sd_b = np.round(sd_b,3)
ts_b = np.round(ts_b,3)
p_values = np.round(p_values,3)
params = np.round(params,4)

df_summary = pd.DataFrame()
df_summary["Coefficients"],df_summary["Standard Errors"],df_summary["t values"],df_summary["Probabilities"] = [params,sd_b,ts_b,p_values]
df_summary.index = ['constant'] + list(X_train.columns)
df_summary
```

### Step7. re-Train Model

```python
# remove column 'SOP'

# step6의 결과를 보고 feature 선택하기(SOP 제거)
# chosen column 이라는 ['GRE Score', 'TOEFL Score', 'University Rating', 'LOR', 'CGPA', 'Research'] 리스트 생성
# X_train, X_test feature 변경

chosen_column = ['GRE Score', 'TOEFL Score', 'University Rating', 'LOR', 'CGPA', 'Research']

X_train = X_train[chosen_column]
X_test = X_test[chosen_column]
```

```python
# from sklearn.linear_model import LinearRegression
# regressor 에 LinearRegression() 오브젝트 할당
# X_train, y_train 으로 regressor fit


from sklearn.linear_model import LinearRegression

regressor = LinearRegression(fit_intercept = True)
regressor.fit(X_train, y_train)
```

```python
# y_pred에 X_test를 regressor으로 predict 한 값을 저장

y_pred = regressor.predict(X_test)
```

```python
plt.scatter(y_test, y_pred, color = 'r')
plt.ylabel('Model Predictions')
plt.xlabel('True (ground truth)')
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, y_pred))

MSE = mean_squared_error(y_test, y_pred)
MAE = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)
MAPE = np.mean( np.abs((y_test - y_pred) / y_test ) ) * 100

print('RMSE =',RMSE, 
      '\nMSE =',MSE, 
      '\nMAE =',MAE, 
      '\nR2 =', r2, 
      '\nAdjusted R2 =', adj_r2, 
      '\nMean Absolute Percentage Error =', MAPE, '%') 
```
